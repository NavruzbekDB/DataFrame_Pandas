```python
import pandas as pd

patient_df = pd.read_csv('data/patient.csv')
'''
 BMI (Body Mass Index)' is also widely used as an indicator of obesity. 
To calculate BMI, divide your weight (kg) by the square of your height (m). 
For example, a person who is 180cm tall and weighs 70kg has a BMI of 21.6, 
which is 70 divided by the square of 1.8.

 BMI (tana massasi indeksi)" ham semirish ko'rsatkichi sifatida keng qo'llaniladi. 
BMI hisoblash uchun vazningizni (kg) bo'yingizning kvadratiga (m) bo'ling. 
Misol uchun, bo'yi 180 sm va vazni 70 kg bo'lgan odamning BMI ko'rsatkichi 21,6 ni tashkil qiladi, 
bu 70 ni 1,8 kvadratga bo'lishdir.
'''
```
